```python
import itertools as it
import numpy as np

def generate_matrices(n):
    # Generate all permutations of 0's and 1's given n even or odd

    if n%2 == 0:
        x = int(n**2/2)
        all_permute = set(it.permutations([0]*x+[1]*x))
    else:
        y = int((n**2-1)/2)
        permute1 = set(it.permutations([0]*y +[1]*(y+1)))
        permute2 = set(it.permutations([0]*(y+1)+[1]*y))

        # Combine the two sets of permutations
        all_permute = permute1.union(permute2)

    unique_matrices = {}

    count = 0

    for i,each in enumerate(all_permute):
        # Reshape the permutation
        matrix = np.array(each).reshape((n, n))
        determinant = np.linalg.det(matrix)

        if determinant == 0:
            count += 1
            unique_matrices.update({count:(determinant, matrix)})

    # print(unique_matrices)

    for i in range(len(unique_matrices)):
        print(f"Matrix {i+1}:")
        print(unique_matrices[i+1][1])
        print("Determinant:", unique_matrices[i+1][0])
        print()


generate_matrices(3)
```
