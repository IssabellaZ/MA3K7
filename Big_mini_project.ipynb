```python
# This code checks the correctness of calculation of lambda_n in the case $N=3$.

def solve(n):
  lambda_0 = 1/6
  lambda_vals = [lambda_0]

  for i in range(1, n+1):
    lambda_n = 1/6 + lambda_vals[-1] * (1/3)
    lambda_vals.append(lambda_n)

  return lambda_vals

# number of iterations n
n = 20
lambda_vals = solve(n)

for i, lambda_n in enumerate(lambda_vals):
    print(f"lambda_{i}: {lambda_n}")
```

```python
import random

# Create a bot that plays ''pub game'' as Beth with an aggressive strategy
# against the user

round_count = 1

def display_glasses(glasses):
    print("Current state of glasses:")
    print(glasses)

def parse_fraction(frac):
    num, deno = map(int, frac.split('/'))
    return num/deno

def ali_turn(glasses):
  global round_count
  distributed_water = 0.5
  print("Round ",round_count, ":")
  while distributed_water >0:
    display_glasses(glasses)

    print("Choose a glass index and water amount to distribute (as a fraction), separate them by a space (q to quit): ")
    user = input()
    if user == 'q':
      return False

    ind, water_str = user.split()
    ind = int(ind)
    amount = parse_fraction(water_str)

    if amount > 1/2:
      print("Too much water to distribute, there isn't enough.")
      continue

    glasses[ind] += amount
    distributed_water -= amount

  return True

def beth_turn(glasses):
  global round_count
  display_glasses(glasses)
  max_index = glasses.index(max(glasses))

  glasses[max_index]=0

  if glasses[max_index+1]>glasses[max_index-1]:
    glasses[max_index+1]=0
    print("Beth chooses to empty glass {} and {}.".format(max_index,(max_index+1)%len(glasses)))
    print("\n")
  elif glasses[max_index+1]<glasses[max_index-1]:
    glasses[max_index-1]=0
    print("Beth chooses to empty glass {} and {}.".format(max_index,(max_index-1)%len(glasses)))
    print("\n")
  else:
    choice = random.choice([-1, 1])
    glasses[max_index + choice] = 0
    print("Beth chooses to empty glass {} and {}.".format(max_index,(max_index+choice)%len(glasses)))
    print("\n")
  round_count += 1

def check_overflow(glasses):
    return max(glasses) > 1

while True:
  num = int(input("Enter the number of glasses on the table: "))
  if num >=1:
    break
  else:
    print("There has to be at least one glass on the table. Please enter 1 or above.\n")
print("\n")
glasses = [0] * num

while True:
  if not ali_turn(glasses):
    print("Game over. You quit.")
    break
  if not check_overflow(glasses):
    beth_turn(glasses)
  if check_overflow(glasses):
    display_glasses(glasses)
    print("Ali wins! One of the glasses overflowed.")
    break
```
